```python
import pandas as pd
import numpy as np
import time
from tensorflow.keras.models import load_model
from sklearn.preprocessing import MinMaxScaler

model = load_model('ANN.keras')  
```

```python
import joblib
scaler_X = MinMaxScaler()
scaler_y = joblib.load('scaler_y.pkl')

data = pd.read_csv('input data.csv')  
data['Time'] = pd.to_datetime(data['Time'])
data['hour'] = data['Time'].dt.hour
data['day_of_week'] = data['Time'].dt.dayofweek
data['month'] = data['Time'].dt.month
data = data.drop(columns=['Time'])
X_scaled = scaler_X.fit_transform(data)
len(data)
```

```python
predictions = []
for i in range(len(data)):
    X_real_time = X_scaled[i].reshape(1, -1)
    predicted_power_scaled = model.predict(X_real_time)
    predicted_power = scaler_y.inverse_transform(predicted_power_scaled)[0][0]
    predictions.append({'Index': i+1, 'Predicted Power(kW)': predicted_power})
    print(f"Index {i+1}: Predicted Power = {predicted_power:.2f}kW")

```
